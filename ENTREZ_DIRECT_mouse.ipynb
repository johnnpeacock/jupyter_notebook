```python
%%bash
esearch -db assembly -query "Mus musculus [ORGN] AND representative [PROP]"| 
elink -target nuccore -name assembly_nuccore_refseq |
#elink -target protein -name gene_protein_refseq |
efetch -format docsum
xtract -pattern DocumentSummary -element AccessionVersion Slen Title | sed 's/,.*//' |
sort -t $'\t' -k 2,2nr
```

```python
%%bash
esearch -db gene -query "Mus musculus [ORGN] AND beta galactosidase [PFN]" | 
elink -target protein -name gene_protein_refseq -cmd neighbor | 
xtract -pattern LinkSet -element Id
```
